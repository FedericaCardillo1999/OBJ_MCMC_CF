```python
import os

# Define your base directory
base_dir = "/Volumes/FedericaCardillo/pre-processing/projects/EGRET+/derivatives/pRFM"

# Define the expected file names (use placeholder {subject})
required_files = [
    "{subject}_ses-01_task-RestingState_hemi-LR_desc-avg_bold_GM.npy",
    "{subject}_ses-01_task-RestingState_hemi-lh_desc-avg_bold_GM.npy",
    "{subject}_ses-01_task-RestingState_hemi-rh_desc-avg_bold_GM.npy",
    "{subject}_ses-02_task-RET_hemi-rh_desc-avg_bold_GM.npy",
    "{subject}_ses-02_task-RET_hemi-lh_desc-avg_bold_GM.npy",
    "{subject}_ses-02_task-RET_hemi-LR_desc-avg_bold_GM.npy",
    "{subject}_ses-02_task-RET2_hemi-rh_desc-avg_bold_GM.npy",
    "{subject}_ses-02_task-RET2_hemi-lh_desc-avg_bold_GM.npy",
    "{subject}_ses-02_task-RET2_hemi-LR_desc-avg_bold_GM.npy"
]

# Loop over each subject
for subject in os.listdir(base_dir):
    if subject.startswith("sub-"):
        subject_path = os.path.join(base_dir, subject)
        if not os.path.isdir(subject_path):
            continue

        print(f"\nChecking files for {subject}...")

        for file_template in required_files:
            file_name = file_template.format(subject=subject)
            session = "ses-01" if "ses-01" in file_name else "ses-02"

            # Paths to check
            nordic_path = os.path.join(subject_path, session, "nordic", file_name)
            nordic_sm4_path = os.path.join(subject_path, session, "nordic_sm4", file_name)

            missing_locations = []
            if not os.path.isfile(nordic_path):
                missing_locations.append("nordic")
            if not os.path.isfile(nordic_sm4_path):
                missing_locations.append("nordic_sm4")

            if missing_locations:
                print(f"  MISSING in {', '.join(missing_locations)}: {file_name}")
```
